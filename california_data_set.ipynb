````python
#import libraries -----
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#loading dataset to variabale called "data"
data = pd.read_excel("/content/housing.xlsx")
````

# 1. What is the average median income of the data set and check the distribution of data using appropriate plots. Please explain the distribution of the plot.



````python
median_house_values = data['median_house_value']
avg_median_house_values = median_house_values.mean()
med_median_house_values = median_house_values.median()
mod_median_house_values = median_house_values.mode()[0]
left_sd_median_house_values = avg_median_house_values - np.percentile(median_house_values, 16)
right_sd_median_house_values = np.percentile(median_house_values, 84) - avg_median_house_values

sns.histplot(median_house_values, color = '#e897da',alpha = 0.3,ec = 'none', kde = True)

plt.axvline(x=avg_median_house_values, linestyle='--', label='Mean' )
plt.axvline(x=med_median_house_values, linestyle='-', label='median' )
plt.axvline(x=mod_median_house_values, linestyle='-.', label='mod' )
plt.axvline(x=left_sd_median_house_values, linestyle=':', label='lef-sd' )
plt.axvline(x=right_sd_median_house_values, linestyle=':', label='right-sd')

sns.rugplot(median_house_values, alpha = 0.025, height=0.05,  label ='individual observations')
plt.xlabel("Average Median House Values")
plt.ylabel('Count')
plt.title('Distribution of Median House Income')
plt.legend( loc ='upper left', fontsize = 7)
plt.show()

print('''
''')
print("null values: ",median_house_values.isnull().sum())
print(f'mode: {median_house_values.mode().tolist()}')
print(f'median: {median_house_values.median()}')
print(f'mean: {median_house_values.mean()}')
print(f'left-sd: {left_sd_median_house_values}')
print(f'right-sd: {right_sd_median_house_values}')
print('''
''')
print(''' \033[39m INSIGHTS:
  1. there are no null values in data
  2. mean > median so that it's "right skewed data or positively skewed data".
   therefore the amount of house values and count of them are inversily proportional to each other 
  3. the rugplot (indicates the individual observations) shows the extream values (outliers)
  4. the last vertical increasing data can be considered as outlier if the data is not necessary
''')
````

# 2. Draw an appropriate plot to see the distribution of housing_median_age and explain your observations.

````python
housing_median_age = data['housing_median_age']

avg_housing_median_age = housing_median_age.mean()
med_housing_median_age = housing_median_age.median()
mod_housing_median_age = housing_median_age.mode()[0]

left_sd_housing_median_age = avg_housing_median_age - np.percentile(housing_median_age, 16)
right_sd_housing_median_age = np.percentile(housing_median_age, 84) - avg_housing_median_age

sns.histplot(housing_median_age, alpha = 0.3,color='#c3acda', ec = 'none', kde = True )

plt.axvline(x=avg_housing_median_age, linestyle='--', label='Mean' )
plt.axvline(x=med_housing_median_age, linestyle='-', label='median' )
plt.axvline(x=mod_housing_median_age, linestyle='-.', label='mod' )
plt.axvline(x=left_sd_housing_median_age, linestyle=':', label='left-sd' )
plt.axvline(x=right_sd_housing_median_age, linestyle=':', label='right_sd' )

sns.rugplot(housing_median_age, alpha = 0.025, height=0.03,  label ='individual observations')

plt.xlabel("median age of houses")
plt.ylabel('Count')
plt.title('Distribution of Median House Income')
plt.legend( loc ='upper left', fontsize = 7)

plt.show()

print('''
''')
print("null values: ",housing_median_age.isnull().sum())
print(f'mode: {housing_median_age.mode().tolist()}')
print(f'median: {housing_median_age.median()}')
print(f'mean: {housing_median_age.mean()}')
print(f'left-sd: {left_sd_housing_median_age}')
print(f'right-sd: {right_sd_housing_median_age}')


print('''
''')
print(''' \033[39m INSIGHTS:
  1. there are no null values
  1. mean < median ---- so that it's "left skewed data or negtively skewed data".
  2. mean   median ---- if not to be precised with float values the data is symetrical (nearly equal).
  3. left and right standard deviation also nearly equal, so it also defines as it is symetrical data
  4. the rugplot (indicates the individual observations) shows the extream values (outliers)
  3. the last vertical increasing data can be considered as outlier if the data is not necessary
''')
````

# 3. Show with the help of visualization, how median_income and median_house_values are related?


The data is continuous, having same size of data and need to be shown co-relation, so scatter plot is best for visualization,

````python
#sns.scatterplot(data = data, x = data['median_income'], y = data['median_house_value'], ec='none', size=8,alpha = 0.3)
fig= plt.figure(figsize=(15, 15))

Relation_plot = sns.jointplot(x=data['median_income'], y=data['median_house_value'],ec='none',alpha = 0.3, data=data, kind='scatter')
Relation_plot.plot_joint(sns.kdeplot, color = '#59c5dc')
Relation_plot.plot_marginals(sns.rugplot, height = -.15, alpha = 0.3 ,color = '#00b0b0', clip_on = False)
plt.axvline(x=data['median_income'].mean(), color='black', linestyle='--', label='Mean' )
plt.axhline(y=data['median_house_value'].median(), linestyle='--', color='black', label='mean' )

plt.xlabel("Median Income")
plt.ylabel("Median House Values")
plt.legend(loc='upper left',fontsize = 5)
#plt.title("RELATIONSHIP BETWEEN MEDIAN INCOME AND MEDIAN HOUSE VALUES")

plt.show()
````

# 10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

````python
def get_bedroom_size(total_bedrooms):
    if total_bedrooms <= 10:
        return 'small'
    elif 11 <= total_bedrooms < 1000:
        return 'medium'
    else:
        return 'large'

data['total_bedroom_size'] = data['total_bedrooms'].apply(get_bedroom_size)
#fig, ax = plt.subplots(figsize=(15, 10))
plt.hist(data['total_bedroom_size'])
plt.show()
````
